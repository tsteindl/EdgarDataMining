```python
path = "data"
```

```python
all_files = glob.glob(os.path.join(path, "*.csv"))

df = pd.concat((pd.read_csv(f, on_bad_lines='skip', sep=";", index_col=False) for f in all_files), ignore_index=True)
```

```python
df["transactionDate"] = pd.to_datetime(df["transactionDate"])
df = df.set_index("transactionDate")
df.sort_index(inplace=True)
df
```

```python
df.plot(use_index=True, figsize=(20, 16))
#plt.figure(figsize=(10, 8))
plt.show()
```
